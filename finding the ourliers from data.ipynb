```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('bigmart.csv')
```

```python
data.head()
```

```python
plt.rcParams['figure.figsize']=(16,4)
plt.subplot(1,4,1)
sns.boxplot(data['Item_Weight'])

plt.subplot(1,4,2)
sns.boxplot(data['Item_Visibility'])

plt.subplot(1,4,3)
sns.boxplot(data['Item_MRP'])

plt.subplot(1,4,4)
sns.boxplot(data['Item_Outlet_Sales'])

plt.suptitle('Checking for univariable outliers')
plt.show()
```

```python
#checking for bivariable outlier wrt target variable

plt.subplot(1,3,1)
sns.scatterplot(y=data['Item_Outlet_Sales'],x=data['Item_Weight'])

plt.subplot(1,3,2)
sns.scatterplot(y=data['Item_Outlet_Sales'],x=data['Item_Visibility'])

plt.subplot(1,3,3)
sns.scatterplot(y=data['Item_Outlet_Sales'],x=data['Item_MRP'])

plt.suptitle('Checking for bivariable wrt target variable')
plt.show()
```

```python
#dealing with the outliers in the real world senerio with bigmart dataset
#capping outliers in the column item outlet sales having values greater than 10000
data['Item_Outlet_Sales'].values[data['Item_Outlet_Sales'].values>10000]=10000
```

```python
#checking if any values greater than 10000 still left in the dataset or not
data[data['Item_Outlet_Sales']>10000]
```

```python
#dealing with the bivariant outlier
#we delete all the values that are greater than 8000
data=data[~(data['Item_Outlet_Sales']>8000)]
```
